````python
def loadme(filename):
    x = []
    for line in open(filename):
        line = line.strip().split(',', 2)
        ng50, ng90, name = line
        ng50 = float(ng50)
        ng90 = float(ng90)
        x.append((ng50, ng90, name))
    return x

iqc = loadme('iqc.ng50.txt')
mqc = loadme('mqc.ng50.txt')
sqc = loadme('sqc.ng50.txt')
````

````python
index_d = {}
index_d_r = {}
labels = []
import nicenames

for n, ( _, _, name) in enumerate(reversed(mqc)):
    nicename = name
    if nicename in labels:
        continue
    index_d[nicename] = n
    index_d_r[n] = nicename
    labels.append(nicename)
    
for n, ( _, _, name) in enumerate(reversed(sqc)):
    nicename = name
    if nicename in labels:
        continue
    index_d[nicename] = n
    index_d_r[n] = nicename
    labels.append(nicename)
    
for n, ( _, _, name) in enumerate(reversed(iqc)):
    nicename = name
    if nicename in labels:
        continue
    index_d[nicename] = n
    index_d_r[n] = nicename
    labels.append(nicename)
    
ng50_iqc = {}
ng90_iqc = {}

for ng50, ng90, name in iqc:
    y = index_d[name]
    ng50_iqc[y] = ng50
    ng90_iqc[y] = ng90
    
ng50_mqc = {}
ng90_mqc = {}

for ng50, ng90, name in mqc:
    y = index_d[name]
    ng50_mqc[y] = ng50
    ng90_mqc[y] = ng90

ng50_sqc = {}
ng90_sqc = {}

for ng50, ng90, name in sqc:
    y = index_d[name]
    ng50_sqc[y] = ng50
    ng90_sqc[y] = ng90

#print(ng50_sqc[45])
#print(ng90_sqc[45])
````

````python
max_y = len(index_d)

iqc_x1 = []
iqc_x2 = []
mqc_x1 = []
mqc_x2 = []
sqc_x1 = []
sqc_x2 = []

megahit_lowest = 0
spades_highest = 0
idba_lowest = 0

for y in range(max_y):
    if y not in ng50_iqc or y not in ng50_mqc or y not in ng50_sqc:
        print('skipping', y, index_d_r[45], y in ng50_iqc, y in ng50_mqc, y in ng50_sqc)
        assert 0
        continue
    iqc_x1.append(ng50_iqc[y])
    iqc_x2.append(ng90_iqc[y])
    
    mqc_x1.append(ng50_mqc[y])
    mqc_x2.append(ng90_mqc[y])

    sqc_x1.append(ng50_sqc[y])
    sqc_x2.append(ng90_sqc[y])
    
    if ng50_mqc[y] <= ng50_iqc[y] and ng50_mqc[y] <= ng50_sqc[y]:
        megahit_lowest += 1

    if ng50_sqc[y] >= ng50_mqc[y] and ng50_sqc[y] >= ng50_iqc[y]:
        spades_highest += 1
        
    if ng50_iqc[y] <= ng50_mqc[y] and ng50_iqc[y] <= ng50_sqc[y]:
        idba_lowest += 1
        
        
max_y = len(mqc_x1)
print('megahit lowest', megahit_lowest)
print('spades highest', spades_highest)
print('idba lowest', idba_lowest)
print(max_y)
````

````python
print(len(mqc_x1), len(sqc_x1), len(iqc_x1))
````

````python
pylab.plot(mqc_x1, range(max_y), '.', label='MEGAHIT')
pylab.plot(iqc_x1, range(max_y), '.', label='IDBA')
pylab.plot(sqc_x1, range(max_y), '.', label='SPAdes')
pylab.xscale('log')
pylab.axis(xmin=100)
pylab.legend(loc='lower left', numpoints=1)
pylab.title('NG50 per genome')

pylab.savefig('ng50.pdf')
````
